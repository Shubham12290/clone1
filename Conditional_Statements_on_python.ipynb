```python
# Match Cases(Python 3.10+)
# just Like Switch Case Features
# Jab ek value ke Multiple Possible cases ho then we perfrom Match Case statement in python

day=int(input("Enter Day Number"))
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case _:
    print("Invalid Input")

```

```python
# ATM Menu
choise =int(input("Enter Your Choise(1,2,3)"))
balance=5000
match choise:
  case 1:
    print("Cash Withdraw")
  case 2:
    print("Deposit")
  case 3:
    print("Check Balance")
    print(f"Your Balance is {balance}")
  case _:# Default Case
    print("Invalid Choise")
```
